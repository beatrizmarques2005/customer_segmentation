## _Customer Basket_

### 0. General

```python
# Explode the list_of_goods column to have one item per row
customer_basket['list_of_goods'] = customer_basket['list_of_goods'].apply(eval)  # Convert string representation of list to actual list
exploded_basket = customer_basket.explode('list_of_goods')

# Group by list_of_goods and count unique invoice_ids and customer_ids
goods_summary = exploded_basket.groupby('list_of_goods').agg(
    invoice_count=('invoice_id', 'nunique'),
    customer_count=('customer_id', 'nunique')
).reset_index()

display(goods_summary)
```

```python
print(f'goods_summary shape: {goods_summary.shape}')
```
